```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
from mpl_toolkits . mplot3d import Axes3D
from numpy.linalg import svd , pinv
```

```python
n =1000

mu1 = np.array([2 ,1 , -3])

mu2 = np.array([1 , -4 ,0])

mu3 = np.array([2 ,4 ,0])

X1 = np.random.randn(n ,3) + mu1

X2 = np.random.randn(n ,3) + mu2

X3 = np.random.randn(n ,3) + mu3

fig = plt.figure()

ax = fig.gca(projection ='3d')

ax.plot(X1[: ,0] , X1[: ,1] , X1[: ,2] , 'r.',alpha =0.5 , markersize =2)

ax.plot(X2[: ,0] , X2[: ,1] , X2[: ,2] , 'b.',alpha =0.5 , markersize =2)

ax.plot(X3[: ,0] , X3[: ,1] , X3[: ,2] , 'g.',alpha =0.5 , markersize =2)

ax.set_xlim3d( -4 ,6)

ax.set_ylim3d( -5 ,5)

ax.set_zlim3d( -5 ,2)

plt.show ()
```

```python
mu21 = (mu2 - mu1).reshape (3 ,1)
mu31 = (mu3 - mu1).reshape (3 ,1)
```

```python
W = np.hstack(( mu21 , mu31))
```

```python
U,_,_ = svd(W) # we only need U
```

```python
n = 1000
```

```python
mu1 = np.array([2 ,1 , -3])
mu2 = np.array([1 , -4 ,0])
mu3 = np.array([2 ,4 ,0])
```

```python
X1 = np.random.randn(n ,3) + mu1
X2 = np.random.randn(n ,3) + mu2
X3 = np.random.randn(n ,3) + mu3
```

```python
fig = plt.figure()
ax = fig.gca(projection ='3d',)
ax.plot(X1[: ,0] , X1[: ,1] , X1[: ,2] , 'r.', alpha = 0.5, markersize =2)
ax.plot(X2[: ,0] , X2[: ,1] , X2[: ,2] , 'b.', alpha = 0.5 , markersize =2)
ax.plot(X3[: ,0] , X3[: ,1] , X3[: ,2] , 'g.',alpha = 0.5, markersize =2)
ax.set_xlim3d(-4, 6)
ax.set_ylim3d(-5, 5)
ax.set_zlim3d(-5, 2)
plt.show()
```

```python
mu21 = (mu2 - mu1).reshape (3 ,1)
mu31 = (mu3 - mu1).reshape (3 ,1)
```

```python
W = np.hstack(( mu21 , mu31))
```

```python
U,_,_ = svd(W) # we only need U
```

```python
P = W @ pinv(W)
```

```python
R = U.T @ P
```

```python
RX1 = (R @ X1.T).T
RX2 = (R @ X2.T).T
RX3 = (R @ X3.T).T
```

```python
plt.plot(RX1 [: ,0] , RX1 [: ,1] ,'b.',alpha =0.5 , markersize =2)
plt.plot(RX2 [: ,0] , RX2 [: ,1] ,'g.',alpha =0.5 , markersize =2)
plt.plot(RX3 [: ,0] , RX3 [: ,1] ,'r.',alpha =0.5 , markersize =2)
plt.show ()
```

### Conclusion
1. The original data is in 3-D and seems quite separable.
2. The dimensionality reduction is performed on this data using SVD and it is converted into 2-D.
3. The newly projected two dimensional data points are better separated than the original three dimensional data points.
4. The dimensionality reduction is achieved and all the necessar information in the data is also retained here.
5. The reatined information and enhanced separability would improve the classification performance.
